```python
#read text file
text=open("C:/Users/DeLL/OneDrive/Documents/Final Assignment Folder - 1/TA_Asgn_1_23030242045/TA_Asgn_1_23030242045/problems.txt").read()
```

```python
print(text)
```

```python
set(stopwords.words('english'))
```

```python
stop_words=set(stopwords.words('english'))
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
text=text.lower()
```

```python
print(text)
```

```python
words=word_tokenize(text)
print(words)
```

```python
#create empty list to save the text without stopwords
text_1=[]
```

```python
for word in words:
    if word not in stop_words:
        text_1.append(word)
print(text_1)
```

```python
#remove punctuation 
punc='''!()-[]{};:'""\,<>./?@#$%^&*_~'''
```

```python
text_2=[]
```

```python
for i in text_1:
    if i not in punc:
        text_2.append(i)
print(text_2)
```

```python
text_3 = ' '.join(map(str, text_2))
 
print(text_3)
```

```python
# Generate a word cloud image
wordcloud = WordCloud(max_font_size=40,colormap='prism',background_color='white').generate(text_3)
```

```python
# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from wordcloud import STOPWORDS
stopwords = set(STOPWORDS)
wordcloud = WordCloud(max_font_size=40,stopwords=mywords, background_color="white",colormap='prism').generate(text_3)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
